```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import numpy.fft as fft
```

```python
import csv
```

```python
t = []
x = []

with open('trace1.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        t.append(float(row[0]))
        x.append(float(row[1]))
```

```python
plt.plot(t,x)
```

```python
plt.plot(np.abs(fft.fft(x)))
```
